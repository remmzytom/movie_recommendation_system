### Data cleaning Phase

```python
#changing all columns to lowercase
df.columns = df.columns.str.lower()
```

```python
# Assuming 'movies' is the column containing movie titles
#df = df.drop_duplicates(subset='movies')
```

```python
#changing from float to string
df['year'] = df['year'].astype(str)
```

```python
#extracting the first 4 digit from each year
df['year'] = df['year'].str.extract(r'(\d{4})')
df.head()
```

```python
df.shape
```

```python
#changing from float to string
df['genre'] = df['genre'].astype(str)
df['genre'] = df['genre'].apply(lambda x: x.replace('\n',''))
df.head()
```

```python
#cleaning the star column 
df['one-line']= df['one-line'].astype(str)
df['stars']= df['stars'].astype(str)
df['one-line']= df['one-line'].apply(lambda x: x.replace('\n', ''))
df['stars']= df['stars'].apply(lambda x: x.replace('\n', ''))
df.head()
```

```python
#seperating the stars name and directors into seperate columns
df['directors'] = df['stars'].str.extract(r'Director:([^|]*)')

# Extract stars
df['stars_names'] = df['stars'].str.extract(r'Stars?:\s*(.*)')

df.head()
```

```python
df.columns
```

```python
#remove the gross column and the stars column
df = df.drop(columns=['stars','gross'])
df.head()
```

```python
#check for all missing values
missing_v = df.isna().sum().sort_values()/len(df)*100
missing_v
```

```python
#handling the missing values in year
df['year'] = df['year'].fillna(df['year'].mode()[0])
df['stars_names'] = df['stars_names'].fillna(df['stars_names'].mode()[0])
df['directors'] = df['directors'].fillna(df['directors'].mode()[0])
df.head()
```

```python
missing_va = df.isna().sum().sort_values()
missing_va
```

```python
#trying to check the distribution of rating,votes,runtime
plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
df['rating'].hist()
plt.title('rating')

plt.subplot(1,3,2)
df['votes'].hist()
plt.title('votes')

plt.subplot(1,3,3)
df['runtime'].hist()
plt.title('runtime')
```

```python
#clean up the comma's in votes column
df['votes'] =df['votes'].str.replace(',', '').astype(float)

#filling in the na with median of the columns
df['rating'] = df['rating'].fillna(df['rating'].median())
df['votes'] = df['votes'].fillna(df['votes'].median())

df['runtime'] = df['runtime'].fillna(df['runtime'].median())
df.head()
```

```python
#making a copy odf the dataframe
new_df = df.copy()
new_df.head()
```

### Genre-Based Analysis
#### Average Rating per Genre:
#### Calculating and visualizing the average rating for each genre. This will show which genres tend to receive higher ratings on average.

```python
new_df['genre_list'] = new_df['genre'].str.lower().str.strip().str.split(', ')
new_df_exploded = new_df.explode('genre_list')
new_df_exploded.head(3)
```

```python
avg_rating_per_genre = new_df_exploded.groupby('genre_list')['rating'].mean().sort_values(ascending=False)

# Step 2: Convert the series to a DataFrame
avg_rating_df = avg_rating_per_genre.reset_index()

# Step 3: Pivot the DataFrame to create a matrix format suitable for heatmap
# In this case, we're using just one dimension (genre) so we can use genre as index and rating as value
heatmap_data = avg_rating_df.pivot_table(values='rating', index='genre_list')

# Step 4: Plot the heatmap
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f", cbar_kws={'label': 'Average Rating'})

# Add labels and title
plt.title('Average Rating per Genre', fontsize=16, fontweight='bold')
plt.ylabel('Genre', fontsize=12)
plt.xlabel('Average Rating', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

Key Observations:
Top-Rated Genres:

- Animation holds the top spot with the highest average rating of 7.33, followed by Adventure at 7.25, and History at 7.26. These genres tend to be well-received by audiences, consistently producing higher-rated films.
- These high average ratings suggest that movies in these genres often appeal to a broad audience or maintain a certain standard of quality.
Moderately Rated Genres:

- Genres such as Action (7.10), Crime (7.08), and Documentary (7.17) fall into the mid-high range, with average ratings just above 7.0. These genres maintain a relatively stable audience and often receive positive feedback, but the variability in quality could lead to a more diverse reception.
Lower-Rated Genres:

- Horror has the lowest average rating at 6.11, followed by Game-Show (6.38) and Thriller (6.45). These genres seem to polarize audiences, with some movies receiving low ratings. Horror in particular is known for having a large number of lower-budget films, which could contribute to the lower average rating.
- Reality-TV (6.65) and Sci-Fi (6.65) also rank on the lower end, suggesting these genres are more divisive or inconsistent in their quality.
Niche Genres:

- Genres like Film-Noir (7.02), Musical (7.02), and Biography (7.06), while not topping the charts, maintain consistent average ratings. These genres cater to more specific audiences and tend to produce stable yet not exceptional ratings.
Balanced Genres:

- Genres like Drama (7.09), Romance (6.83), and Mystery (7.09) have well-balanced average ratings, hovering just above the 7.0 mark. These genres likely have a wide variety of films ranging from critically acclaimed to average, resulting in consistent but moderate ratings overall.

Insights by Genre:

### Clustering Movies
### This could reveal patterns or clusters of movies that tend to be similar (e.g., action-packed, highly-rated thrillers)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Split each genre, strip whitespace, and rejoin
new_df['genre'] = new_df['genre'].apply(lambda x: ','.join([genre.strip() for genre in x.split(',')]))

# One-hot encoding: converting genres to dummy variables
df_genres = new_df['genre'].str.get_dummies(sep=',')

# Combine with the main dataset
features = pd.concat([new_df[['votes', 'runtime', 'rating']], df_genres], axis=1)

# Scaling numerical features
scaler = StandardScaler()
features[['votes', 'runtime', 'rating']] = scaler.fit_transform(features[['votes', 'runtime', 'rating']])
```

```python
features.head()
```

```python
features.columns
```

##### using the elbow method

```python
from yellowbrick.cluster import KElbowVisualizer

import warnings
# Suppress specific UserWarnings related to KMeans and MKL
warnings.filterwarnings("ignore", message="KMeans is known to have a memory leak on Windows with MKL, when there are less chunks than available threads.")
import os
os.environ["OMP_NUM_THREADS"] = "2"

model = KMeans(n_init=10)
visualizer = KElbowVisualizer(
    model,  k=(2,12), metric='distortion', timings=False
) #distortion same as Euclidean distance

visualizer.fit(features)
```

```python
# Assume the elbow plot suggests 4 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
cluster_label = kmeans.fit_predict(features)  # Assign clusters to each movie
cluster_label
```

##### Applying PCA

```python
from sklearn.decomposition import PCA
p_ca = PCA(n_components = 2)
pca_features = p_ca.fit_transform(features)

#convert the PCA into dataframe for plotting
pca_df = pd.DataFrame(data=pca_features, columns=['PCA1','PCA2'])
pca_df['cluster'] = cluster_label  #add the cluster labels

plt.figure(figsize=(10,6))
sns.scatterplot(x='PCA1',y='PCA2', hue='cluster', data=pca_df, palette='Set1', s=100, alpha=0.7)
plt.title('K-mean clustering of Movies')
plt.xlabel('principal component 1')
plt.ylabel('principal component 2')
plt.legend(title='cluster')
plt.show()
```

This plot shows how movies are grouped into five different clusters based on their characteristics. Each dot represents a movie, and the colors indicate the cluster each movie belongs to.

- Distinct Groups: The orange group (Cluster 4) stands out separately, meaning these movies likely have unique traits compared to the others.
- Overlap in Groups: Some groups, like red, blue, and purple, are close to each other, suggesting they share some similar features.
- Usefulness: These clusters help us understand the types of movies in each group, which can be useful for recommending similar movies to viewers or understanding different movie styles.

```python
# Assuming you have your original dataframe 'df' and 'kmeans' is your KMeans model
# Add the cluster labels to your original dataframe
new_df['cluster'] = cluster_label
new_df['cluster']
```

```python
# Group by cluster and calculate mean for numerical features
cluster_summary = new_df.groupby('cluster').agg({
    'votes': 'mean',
    'runtime': 'mean',
    'rating': 'mean'
}).rename(columns={
    'votes': 'avg_votes',
    'runtime': 'avg_runtime',
    'rating': 'avg_rating'
})

# Display the summary statistics
print("Cluster Summary:\n", cluster_summary)
```

```python

df_genres['cluster'] = new_df['cluster']  # Add cluster labels to the genre dummy DataFrame

# Calculate genre distribution for each cluster
genre_distribution = df_genres.groupby('cluster').sum()

# Normalize to get percentages for each genre within each cluster
genre_distribution = genre_distribution.div(genre_distribution.sum(axis=1), axis=0) * 100

print("Genre Distribution by Cluster:\n", genre_distribution)
genre_distribution.head()
```

```python
# Define an empty list to store cluster profiles
cluster_profiles = []

# Loop over each cluster and summarize characteristics
for cluster_id in cluster_summary.index:
    # Extract cluster summary stats
    avg_votes = cluster_summary.loc[cluster_id, 'avg_votes']
    avg_runtime = cluster_summary.loc[cluster_id, 'avg_runtime']
    avg_rating = cluster_summary.loc[cluster_id, 'avg_rating']
    
    # Extract the dominant genres for this cluster
    dominant_genres = genre_distribution.loc[cluster_id].nlargest(4).index.tolist()
    dominant_genres = ", ".join(dominant_genres)
    
    # Define an audience type based on the characteristics
    if avg_rating >= 7.5:
        audience_type = "High-rating audience"
    elif avg_rating < 6.0:
        audience_type = "Casual viewers"
    else:
        audience_type = "General audience"

    # Append profile data as a dictionary
    cluster_profiles.append({
        'Cluster': cluster_id,
        'Avg Rating': round(avg_rating, 2),
        'Avg Runtime (mins)': round(avg_runtime, 2),
        'Avg Votes': round(avg_votes),
        'Dominant Genres': dominant_genres,
        'Audience Type': audience_type
    })

# Convert the list of dictionaries to a DataFrame for display
profile_df = pd.DataFrame(cluster_profiles)
print(profile_df)
profile_df
```

### Recommendation System  

```python
#craeting a  new column to the new_df df calling it tags

new_df['tags'] = new_df['genre'] + ' ' + new_df['one-line'] + ' ' + new_df['directors'] + ' ' + new_df['stars_names']
new_df.head()
```

```python
new_features = new_df.copy()
new_features.head()
```

```python
#using TF-IDF to  transform the tags into numerical data

from sklearn.feature_extraction.text import TfidfVectorizer

tdf = TfidfVectorizer(stop_words = 'english')
tdf_mat = tdf.fit_transform(new_features['tags'])
```

```python
#computing the consine similarities
from sklearn.metrics.pairwise import cosine_similarity

cos_sim = cosine_similarity(tdf_mat, tdf_mat)

cos_sim
```

#### building Recommendation function

```python
indices = pd.Series(new_features.index, index=new_features['movies']).drop_duplicates()

def recommend_movies(title, cos_sim=cos_sim, year_filter=None,rating_filter=None):
    #getting the the index of the movies that matches the title
    idx = indices[title]
    # Get the similarity scores of all movies with the target movie
    sim_scores = list(enumerate(cos_sim[idx]))
    # Sort movies based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
     # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]
    #the movie indices
    movie_indices = [i[0] for i in sim_scores]
    #return the top 10 most similar movies
    similar_movies = new_features.iloc[movie_indices]
    # Apply year filter if provided
    if year_filter:
        similar_movies = similar_movies[similar_movies['year'] >= year_filter]

    if rating_filter:
        similar_movies = similar_movies[similar_movies['rating'] >= rating_filter]

    return similar_movies[['movies','year', 'genre','rating']].head(10)
```

```python
#movie recommendation withot filtering
#recommend_movies('Blood Red Sky')

#movie recommendation with filtering
recommend_movies(
    'The Walking Dead',
    year_filter=2010,  # Only movies released after 2010
    rating_filter=7.0,  # Only movies with a rating above 7.0
)
```
